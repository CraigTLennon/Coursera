Grouping data
the groupby operation splits, analyzes and aggregates/combines data


```python
import pandas as pd
import numpy as np
census = pd.read_csv('census.csv')
census.head()
cst=census.groupby('STNAME')  #this produces a broupby object, with the column on which we used groupby as the key.  we can 
#see the groups by
cst.groups # where the value is a list of indices belonging to that group
#we could see which states have the most counties by using size (count counties) and sort( get top to list)
cst.size().sort_values(ascending = False)
#to see the biggest counties we could sort by county and state
cty=census.groupby(['STNAME','CTYNAME'])
#for st, cty in cty.groups:
    #print( st, cty)
cty['POPESTIMATE2010'].sum().sort_values(ascending=False) 
#I could go back and remove the state aggregates, using '50' but do not need to
cty=census.groupby('SUMLEV')
cty.agg([np.sum,np.mean,np.size])
```

groupby also has a transform method and a filtering method to modify or get subsets of groupby objects

```python
cty=census.groupby(['STNAME','CTYNAME'])
cty.filter(lambda x: np.all([len(x['STNAME'])>1 ]))
```

For mergind and joining data structures, you use concat and append
concat joins multiple data structures along a single axis
append joins only along axis 0, row axis

```python
import pandas as pd
import numpy as np
census = pd.read_csv('census.csv')
census.iloc[0:5]
pd.concat([census.iloc[0:3],census.iloc[3:6]],axis=0)
pd.concat([census.iloc[:,0:3],census.iloc[:,3:6]],axis=1)
census.iloc[0:3].append(census.iloc[3:6])
```
